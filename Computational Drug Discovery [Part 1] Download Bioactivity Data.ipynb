```python
# Importing libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
# pd.set_option('display.max_columns', None)
```

```python
#Trget search for coronavirus
target = new_client.target
target_query= target.search('coronavirus')
targets= pd.DataFrame.from_dict(target_query)
targets
# targets.dtypes
```

```python
#Select and retrieve bioactivity data for Severe acute respiratory syndrome coronavirus 2  (tenth entry)
selected_target= targets.target_chembl_id[10]
selected_target
```

```python
#here we will retrieve only bioactivity data (IC50) for CHEMBL4523582 that are reported as IC500 values in nM
activity = new_client.activity
res1= activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
#convertig the res in a dataframe
df= pd.DataFrame.from_dict(res1)
# df.head()
# df.standard_type.unique()
```

```python
#Handling missing data
#If any compounds has missing value for the standard_value and canonical_smiles column then drop it.
df2= df[df.standard_value.notna()]
df2= df[df.canonical_smiles.notna()]
df2
```

```python
#dropping duplicates from canonical smiles
df2_nr= df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
#Combine the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and bioactivity_class into a DataFrame
#as we are interested only in IC50
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3= df2_nr[selection]
df3
```

```python
#saving df3 as CSV file
df3.to_csv('bioactivity_data_preprocessed.csv', index= False)
```

#**Labeling compounds as either being active, inactive or intermediate**
1. Compounds having values of less than 1000 nM will be considered to be active
2. Compounds having values of more than 10,000 nM will be considered to be inactive
3. Compounds having values between 1000 and 10,000 nM will be considered to be intermediate

```python
#Read the saved bioactivity_data csv file into a dataframe
df4= pd.read_csv('bioactivity_data_preprocessed.csv')
df4.head()
df4.shape #to see the total columns and rows
df4.describe()
```

```python
#labeling compounds according to their bioactivity
#First create a empty list
bioactivity_threshold= []
for i in df4.standard_value:
  if float(i)>10000:
    bioactivity_threshold.append('inactive')
  elif float(i)<1000:
    bioactivity_threshold.append('active')
  else:
    bioactivity_threshold.append('intermediate')
bioactivity_class= pd.Series(bioactivity_threshold, name= 'class')
# print(bioactivity_class)
df5= pd.concat([df4,bioactivity_class],axis=1)
df5
```

```python
#save the df5 into a CSV
df5.to_csv('Bioactivity_data_curated.csv',index=False)
```
